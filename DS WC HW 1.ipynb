## Диплом на тему: аналитика динамики продаж квартир.

# 1. Бизнес-анализ (Business understanding)

### Определение бизнес-целей

- Предсказать вероятность продажи квартиры (на конкретную дату) и оценить количественно стоимость.

### Аналитическая постановка

- Метрики для оценки вероятности можно использовать, например, ROC-AUC. Для оценки количественной стоимости: RMSE, Logloss или MSE.

# 2. Анализ данных (Data understanding)

### Сбор данных

- Данные будут предоставлены компанией в готовом виде, поэтому на сбор тратить время не придется.

### Описание данных

- В данном этапе считаем основные статистики и метрики по данным.

### Исследование данных

- Визуализируем данные, ищем оптимальные способы наглядно показать закономерности в данных.

### Качество данных

- Рассматриваем данные более пристально, ищем пропущенные значение, ошибки. Выбираем стратегию борбы с ними, например, заполнение пропуски средним, медианой и тд.

**Т.о. на второй блок можно затратить от около двух часов, если данные качественные, все колонки в данных имеют прямое отношение к задаче и их связь легко считывается на визуализации и до нескольких дней, если данные крайне некачественные и неполные (требуют некоторого дополнительно сбора и поиска).**

# 3. Data Preparation (подготовка данных)

### Отбор данных

- Отбор релевантных и качественных атрибутов с низкой корреляцией между собой.

### Генерация данных

- Агригируем данные, а также нормализуем и подготавливаем для нашей модели.

**В целом, если это часть анализа не должна занять большое количество, вероятно за несколько часов можно управиться с подготовкой данных.**

# 4. Modeling (Моделирование)

### Выбор алгоритмов

- Для начала можно использовать обычную линейную и логистическую регрессию и затем пробовать более сложные модели, а возможно и ансамбли, если потребуется еще большая точность.

### Планирование тестирования.

- Используем gridsearch для оптимизации гиперпараметров наших моделей. 
- Используем перекрестную валидацию для борьбы с переобучением. 

### Оценка результатов

- оцениваем результаты на основе наших метрик и конвертируем это в более понятные для бизнеса метрики.

**Этот этап может также занять от нескольких часов, если данных не сильно много и все алгоритмы отработают быстро до несколько дней, если имеются большое количество данных, а также потребуется создание сложных моделей, например, ансамблей для повышения точности.**

# 5. Отчет по результатам моделирования

- написание фильнального отчета и презентации по проделанной работе.


**Может занят от одного дня времени.**
